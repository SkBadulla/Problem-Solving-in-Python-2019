```python
# numbers vs cubes
import matplotlib.pyplot as plt
fig = plt.figure()
n = list(range(10))
sqn = [i ** 10 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
# Voltage - Time
import numpy as np
# t = [i/10 for i in range(1,21)]
t = np.arange(0.1,2.0,0.1)
v = [1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Volatege Graph')
# plt.plot(t,v,'go')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')

plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('DataFiles/income.csv')
years = df.columns[2:]

incomeArkansa = df.values[3,2:] 
plt.title("Income of Arkansas")
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansa)
plt.show()
```

```python
# ncome of All States from 2005 to  2013
years=df.columns[2:]
#incomeallstates = [sum(df[year])/ for year in years]

incomeallstates = [df[year].mean() for year in years]
plt.title("Income of allStates")
plt.xlabel('Year')
plt.ylabel('incomeallstates')
plt.plot(years,incomeallstates)
plt.show()
```
